```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
data.drop('name', axis=1, inplace=True)
data.drop('userid', axis=1, inplace=True)
data.drop('name.1', axis=1, inplace=True)
data.drop('userid.1', axis=1, inplace=True)
data.drop('name.2', axis=1, inplace=True)
data.drop('userid.2', axis=1, inplace=True)
```

```python
# data['victory'] = np.random.randint(0,2,len(data))
```

```python
data.head()
```

```python
data.describe()
```

```python
X = data.drop("victory", axis=1)
y = data["victory"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```

```python
from sklearn.metrics import classification_report

y_pred = classifier.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
gb.score(X_test, y_test)
```
